```python
# ========================= FUNCAO DE PLOTS 
import matplotlib.pyplot as plt

def rms_plot(s1, s1l, s2, s2l, s3, s3l, s4, s4l, nome, caminho_salvar,
             ymin=None, ymax=None, figout=None, xticks=None,
             title='RMSE 2018-2022 - componentes u/v do vento - AMV-DSA (IR) x Reanalise ERA5'):
    
    plt.figure(figsize=(12, 3))
    plt.plot(xticks, s1, label=s1l, color='red')
    if s2 is not None and not s2.empty:
        plt.plot(xticks, s2, label=s2l, color='green')
    if s3 is not None and not s3.empty:
        plt.plot(xticks, s3, label=s3l, color='red', ls='--')
    if s4 is not None and not s4.empty:
        plt.plot(xticks, s4, label=s4l, color='green', ls='--')
    
    plt.title(title)
    plt.ylabel('m/s')
    plt.ylim(ymin, ymax)
    plt.xlabel('months')
    plt.xticks(rotation=90)
    plt.legend(loc='best')

    if figout:
        plt.savefig(f'{caminho_salvar}/{figout}_{nome}.png', dpi=100, bbox_inches='tight')
    else:
        plt.show()
def plot_rmse_bias_altura(var, press_min, press_max,
                          data1_x,data1_y,nome_data1,
                           data2_x,data2_y,nome_data2,
                            data3_x,data3_y,nome_data3,
                             data4_x,data4_y,nome_data4,
                              data5_x,data5_y,nome_data5,
                               data6_x,data6_y,nome_data6,
                               data7_x,data7_y,nome_data7,
                               data8_x,data8_y,nome_data8,
                                data1_y2,
                                data2_y2,
                                data3_y2,
                                data4_y2,
                                data5_y2,
                                data6_y2,
                                data7_y2,
                                data8_y2, caminho_salvar_imagem                          
                                ):
    # ============================ PLOTS 
    #======plot 1
    fig, axs = plt.subplots(2,figsize= (12,6),  gridspec_kw={'hspace': 0.7} )

    #plt.figure(figsize=(12, 3))
    axs[0].plot(data1_x, data1_y,  label=nome_data1, color='red')
    axs[0].plot(data2_x, data2_y,  label=nome_data2, color='blue')
    axs[0].plot(data3_x, data3_y,  label=nome_data3, color='pink')
    axs[0].plot(data4_x, data4_y,  label=nome_data4, color='purple')
    axs[0].plot(data5_x, data5_y,  label=nome_data5, color='brown')
    axs[0].plot(data6_x, data6_y,  label=nome_data6, color='black')
    axs[0].plot(data7_x, data7_y,  label=nome_data7, color='yellow')
    axs[0].plot(data8_x, data8_y,  label=nome_data8, color='gray')

    axs[0].set_title(f'Comparação {var} componente U do vento de {press_min} a {press_max} hpa')
    axs[0].set_ylabel('m/s')
    axs[0].set_ylim(-5, 5)
    axs[0].tick_params(axis='x', rotation=70)

    # PLOT 2 
    axs[1].plot(data1_x, data1_y2,  label=nome_data1, color='red')
    axs[1].plot(data2_x, data2_y2,  label=nome_data2, color='blue')
    axs[1].plot(data3_x, data3_y2,  label=nome_data3, color='pink')
    axs[1].plot(data4_x, data4_y2,  label=nome_data4, color='purple')
    axs[1].plot(data5_x, data5_y2,  label=nome_data5, color='brown')
    axs[1].plot(data6_x, data6_y2,  label=nome_data6, color='black')
    axs[1].plot(data7_x, data7_y2,  label=nome_data7, color='yellow')
    axs[1].plot(data8_x, data8_y2,  label=nome_data8, color='gray')
    # SETANDO TITULO
    axs[1].set_title(f'Comparação {var} componente V do vento de {press_min} a {press_max} hpa')
    axs[1].set_ylabel('m/s')
    axs[1].tick_params(axis='x', rotation=70)
    axs[1].set_ylim(-5, 5)
    axs[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=5)
    # ------- SALVANDO PLOT -----------------
    plt.savefig(f'{caminho_salvar_imagem}/{var}_{press_min}_{press_max}hpa.png', dpi=100, bbox_inches='tight')
def plot_NOBS(var, press_min, press_max,
                          data1_x,data1_y,nome_data1,
                           data2_x,data2_y,nome_data2,
                            data3_x,data3_y,nome_data3,
                             data4_x,data4_y,nome_data4,
                              data5_x,data5_y,nome_data5,
                               data6_x,data6_y,nome_data6,
                               data7_x,data7_y,nome_data7,
                               data8_x,data8_y,nome_data8,
                               caminho_salvar_imagem                          
                                ):
    # ============================ PLOTS 
    #======plot 1
    #fig, axs = plt.subplots(2,figsize= (12,6),  gridspec_kw={'hspace': 0.7} )

    plt.figure(figsize=(12, 3))
    plt.plot(data1_x, data1_y,  label=nome_data1, color='red')
    plt.plot(data2_x, data2_y,  label=nome_data2, color='blue')
    plt.plot(data3_x, data3_y,  label=nome_data3, color='pink')
    plt.plot(data4_x, data4_y,  label=nome_data4, color='purple')
    plt.plot(data5_x, data5_y,  label=nome_data5, color='brown')
    plt.plot(data6_x, data6_y,  label=nome_data6, color='black')
    plt.plot(data7_x, data7_y,  label=nome_data7, color='yellow')
    plt.plot(data8_x, data8_y,  label=nome_data8, color='gray')

    plt.title(f'Comparação {var}-{press_min} a {press_max} hpa')
    plt.ylabel('')
    plt.xticks(rotation=70)
    
    plt.savefig(f'{caminho_salvar_imagem}/{var}_{press_min}_{press_max}hpa.png', dpi=100, bbox_inches='tight')

# ========================================
```
